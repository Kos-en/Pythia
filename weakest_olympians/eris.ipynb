```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Download and prepare the data
dataroot = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(dataroot + "lifesat/lifesat.csv")
X = lifesat[["GDP per capita (USD)"]].values
Y = lifesat[["Life satisfaction"]].values
```

```python
# Plot the data
lifesat.plot(kind="scatter",grid=True, x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23600,63000,4,9])
plt.show()
```

```python
# Select a model
model = LinearRegression()
model_knn = KNeighborsRegressor(n_neighbors=3)
```

```python
# Train the linear model
model.fit(X,Y)
```

```python
# Train the KNN model
model_knn.fit(X,Y)
```

```python
# Make a prediction for Puerto Rico
X_puerto_rico = [[33442.8]]
print(model.predict(X_puerto_rico))
```

```python
X_puerto_rico = [[33442.8]]
print(model_knn.predict(X_puerto_rico))
```
